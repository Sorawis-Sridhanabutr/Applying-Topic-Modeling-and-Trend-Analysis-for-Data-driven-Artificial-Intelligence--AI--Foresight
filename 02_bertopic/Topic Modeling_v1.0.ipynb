```python
import pandas as pd
df = pd.read_csv("/Users/moo/Desktop/MSc Business Analytics/BEMM466 Business project/BERTopic Finalised/arxiv_ai_cleaned_ready.csv")
df.head(10)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np


docs = df['summary_clean'].astype(str).tolist()
# sbert_model = SentenceTransformer('all-mpnet-base-v2')
# embeddings = sbert_model.encode(docs, show_progress_bar=True)

# np.save("embeddings.npy", embeddings)

embeddings = np.load("embeddings.npy")
```

```python
from umap import UMAP
from hdbscan import HDBSCAN
from bertopic import BERTopic

umap_model = UMAP(n_neighbors=15, n_components=5, min_dist=0.0, metric="cosine")
hdbscan_model = HDBSCAN(min_cluster_size=25, min_samples = 10, metric="euclidean", prediction_data=True)

topic_model = BERTopic(
    embedding_model=None,
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    calculate_probabilities=True,
    min_topic_size=10,
    verbose=True
)
```

```python
topics, probs = topic_model.fit_transform(docs, embeddings)
```

```python
topic_info = topic_model.get_topic_info()

# Total number of documents
total_docs = len(docs)

# Number of valid topics (excluding -1)
n_topics = (topic_info['Topic'] != -1).sum()

# Number of documents in noise
noise_count = topic_info[topic_info['Topic'] == -1]['Count'].values[0]

# Noise percentage
noise_pct = (noise_count / total_docs) * 100

print(f"Total Topics (excluding noise): {n_topics}")
print(f"Noise Documents (Topic -1): {noise_count} ({noise_pct:.2f}%)")
```

```python
from gensim.models import CoherenceModel
from gensim.corpora import Dictionary

# Tokenized documents
tokenized_docs = [doc.split() for doc in docs]

# Create dictionary
dictionary = Dictionary(tokenized_docs)

# Top 10 words per topic
topics = topic_model.get_topics()
top_words = [[word for word, _ in topics[i][:10]] for i in topics if i != -1]

# Coherence calculation
coh_model = CoherenceModel(
    topics=top_words,
    texts=tokenized_docs,
    dictionary=dictionary,
    coherence='c_v'
)
coherence_score = coh_model.get_coherence()
print(f"Gensim Coherence (c_v): {coherence_score:.4f}")
```

```python
topics = topic_model.get_topics()
top_n = 10  # number of words per topic to consider

all_words = []
for topic in topics.values():
    if topic:
        all_words.extend([word for word, _ in topic[:top_n]])

unique_words = set(all_words)
diversity_score = len(unique_words) / len(all_words)

print(f"Topic Diversity: {diversity_score:.4f}")
```
